```python
# Create a Tic Tac Toe game
def symbol_selection():
    symbol_p1 = 'Wrong'
    
    while symbol_p1 not in ['X','O']:
        symbol_p1 = input("Player 1: Do you want to be X or O?")
        if symbol_p1 not in ['X','O']:
            print("Your input is invailed, Please select 'X' or 'O'!")
        else:
            print("Player 1 will go first.")
    return symbol_p1
    
    
```

```python
def looping_for_input():
    choice_p1 = -1
    while choice_p1 not in range(0,9):
        choice_p1 = int(input("Choose your number between 0 to 8"))
        if choice_p1 not in range(0,9):
            print("Sorry invailed input")
    return choice_p1
```

```python
def board_display(my_list):
    print("   |   |\n",my_list[0],"|",my_list[1],"|",my_list[2],
                      "\n   |   |   \n-----------\n   |   |\n",my_list[3],
                      "|",my_list[4],"|",my_list[5],"\n   |   |   \n-----------\n   |   |\n",
                      my_list[6],"|",my_list[7],"|",my_list[8],"\n   |   |")
```

```python
def check_board(my_list):
    symbol_list = ['O','X']
    for i in symbol_list:
        if my_list[0]==i and my_list[1]==i and my_list[2]==i:
            return True
        elif my_list[3]==i and my_list[4]==i and my_list[5]==i:
            return True
        elif my_list[6]==i and my_list[7]==i and my_list[8]==i:
            return True
        elif my_list[0]==i and my_list[3]==i and my_list[6]==i:
            return True
        elif my_list[1]==i and my_list[4]==i and my_list[7]==i:
            return True
        elif my_list[2]==i and my_list[5]==i and my_list[8]==i:
            return True
        elif my_list[0]==i and my_list[4]==i and my_list[8]==i:
            return True
        elif my_list[2]==i and my_list[4]==i and my_list[6]==i:
            return True
    else:
        return False
```

```python
def replay():
    from IPython.display import clear_output
    replay = 'Wrong'
    while replay not in ['Yes','No']:
        replay = input("Do you want to play again? Type : Yes or No")
        if replay not in ['Yes','No']:
            print("Please enter 'Yes' or 'No'")
        elif replay == 'No':
            print("Thankyou!")
        else:
            clear_output()
            initial_call()
```

```python
def lets_play(symbol_p1):
    my_list = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    board_display(my_list)
    final_output = False
    empty_check = False
          
    while final_output == False:
             # Check any element in the list is empty or not
        for i in my_list:
            i.strip()
            if not i or i.isspace():
                empty_check = True
                break
            else:
                empty_check = False
        if empty_check == True:
        # Get user input Value
            position = looping_for_input()
            if my_list[position] == ' ':
                my_list[position] = symbol_p1
                board_display(my_list)
                checkval = check_board(my_list)
                if checkval == False:
                    if symbol_p1 == 'X':
                        symbol_p1 = 'O'
                    else:
                        symbol_p1 = 'X'
                else:
                    print("Congratulations you have won!")
                    final_output == True
                    replay()
                    break
            else:
                print("Value is already added at the position. Please enter another number")
        else:
            print("No one won!")
            final_output = True
            replay()
```

```python
def initial_call():
    print("Welcome to Tic Tac Toe !")
    player1_symbol = symbol_selection()
    ready = 'Wrong'
    while ready not in ['Yes','No']:
        ready = input("Are you ready to play?\nEnter Yes or No")
        if ready not in ['Yes','No']:
            print("Please enter 'Yes' or 'No'")
        elif ready == 'No':
            print("Thankyou!")
            break
        else:
            lets_play(player1_symbol)
```
